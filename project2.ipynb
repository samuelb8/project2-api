## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed
url = 'https://en.wikipedia.org/wiki/University_of_Illinois_Urbana-Champaign'
def get_text(url):
    # Get the page content
    response = requests.get(url)
    soup = bs4.BeautifulSoup(response.text, 'html.parser')
    
    # Extract text
    text = soup.get_text()
    
    # Save to file
    with open('output.txt', 'w', encoding='utf-8') as f:
        f.write(text)
    
    print("Text content saved to output.txt")
# Call the function
get_text(url)
# Read the text file
with open('output.txt', 'r', encoding='utf-8') as f:
    text = f.read()
# Print the first 500 characters of the text
print(text)
```

```python
#download an image with beautifulsoup and save it in this repository
url = 'https://en.wikipedia.org/wiki/University_of_Illinois_Urbana-Champaign'
def download_image(url):
    # Get the page content
    response = requests.get(url)
    soup = bs4.BeautifulSoup(response.text, 'html.parser')
    
    # Find the first image
    img_tag = soup.find('img')
    
    if img_tag and img_tag.get("src"):
        img_url = img_tag["src"]
        
        # Check if the image URL is relative
        if img_url.startswith("//"):
            img_url = "https:" + img_url
        elif img_url.startswith("/"):
            img_url = "https://en.wikipedia.org" + img_url
        
        img_response = requests.get(img_url)

        with open('image.jpg', 'wb') as img_file:
            img_file.write(img_response.content)
        
        print("Image downloaded and saved as image.jpg")
    else:
        print("No image found on the page.")

download_image(url)

```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links
url = 'https://en.wikipedia.org/wiki/University_of_Illinois_Urbana-Champaign'
def find_links(url):
    response = requests.get(url)
    soup = bs4.BeautifulSoup(response.text, 'html.parser')
    
    links = soup.find_all('a')
    
    for link in links[:10]:
        print(link.get('href'))

find_links(url)
#print the first 100 characters of ten of these tables
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"
def add_text_column(scripts):
    texts = []
    for index, row in scripts.iterrows():
        try:
            with open(f'script_{index}.txt', 'r', encoding='utf-8') as f:
                text = f.read()
            texts.append(text)
        except FileNotFoundError:
            texts.append("")  # Append empty string if file not found
    
    scripts['text'] = texts
    scripts.to_csv('pudding_texts.csv', index=False)
    print("New DataFrame saved as pudding_texts.csv")
# Call the function
add_text_column(scripts)
```

#### Browse the documentation at https://developer.themoviedb.org/reference/intro/getting-started. Create an account to authenticate

```python
%pip install tmdbsimple
import tmdbsimple as tmdb

tmdb.API_KEY = ''
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 
api_key = 
url = 'https://api.themoviedb.org/3/movie/now_playing'
def get_movies(api_key):
    params = {
        'api_key': api_key,
        'language': 'en-US',
        'page': 1
    }
    response = requests.get(url, params=params)
    data = response.json()
    movies = data['results']
    movie_data = [] # List to store movie data
    for movie in movies:
        movie_info = {
            'title': movie['title'],    
            'release_date': movie['release_date'],
            'overview': movie['overview'],
            'vote_average': movie['vote_average'],
            'popularity': movie['popularity']
        }
        movie_data.append(movie_info)
    return movie_data

# Call the function
movies = get_movies(api_key)

print(movies)
```
